```python
%pip install pandas
%pip install openpyxl
# %pip install shutil
# %pip install os
# %pip install glob
# %pip install datetime

import pandas as pd
import shutil
import os
from glob import glob
from datetime import datetime

# Define volume paths (update as needed)
input_dir = "/Volumes/sales_data_project/sales_data/source_data"
processed_dir = "/Volumes/sales_data_project/sales_data/source_data_csv"
archive_dir = "/Volumes/sales_data_project/sales_data/source_data_processed"

# Current timestamp for filenames
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

# Find all Excel files
excel_files = glob(os.path.join(input_dir, "*.xlsx"))

if not excel_files:
    print("No Excel files found in the input directory")
else:
    for excel_file in excel_files:
        try:
            base_name = os.path.splitext(os.path.basename(excel_file))[0]
            print(f"Processing: {base_name}.xlsx")

            # Read Excel file
            df = pd.read_excel(excel_file)

            # Write CSV with timestamp
            csv_path = os.path.join(processed_dir, f"{base_name}_{timestamp}.csv")
            df.to_csv(csv_path, index=False)

            # Move and rename original Excel file to archive
            archived_path = os.path.join(archive_dir, f"{base_name}_{timestamp}.xlsx")
            shutil.move(excel_file, archived_path)

            # print(f"‚úÖ Converted ‚Üí {csv_path}")
            # print(f"üì¶ Archived ‚Üí {archived_path}")

        except Exception as e:
            print(f"‚ùå Error processing {excel_file}: {e}")

print("üéâ Done.")
```
